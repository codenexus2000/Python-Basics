Panda Dataframes:
-Dataframes are 2D datastructures having data aligned in tabular format
-Data gets aligned in rows called indexes
-Written as - pd.DataFrames(data=[] , index=[], coulmns=[])
-In a dataframe the columns can be named 

Importing Dataframes:
-In pandas u can import csv,json,excel,text,etc. files
-You can load the files by providing url from the web
-Use the command: pd.read_csv('')
-if a file has sepration by specical characters such as , ' ; use sep=''

*Creating a new column in dataframe you simply call a new column as if it was existing

* inplace = True, will save the activity at the time instantly any changes will be saved

Merging:
- pd.merge(left='' , right='', on='', how='')
-the left column has the left dataframe
-the right column has the left dataframe
-on represents on which column the merge is supposed to be perform. (The common column in a dataframe)
-how has two values INNER or OUTER. Inner will represent only the common values. Outer will represent all the values
*In How you can also mention the left and right dataframe which will print the whole data frame only up to the mentioned data frame

- pd.head() gives first 5 rows
- pd.tail() gives last 5 rows
- pd.sample() gives random rows
- pd.groupby() can group different categories in a column
- pd.xs(key= , level= ) can access data 
- pd.plot.box() will plot a box plot of the data
- pd.describe() will provide count, mean, standard deviation, minimum, percentages (25,50,75), maximum for each row
- pd.transpose makes the rows into column and column into rows
- pd.unique() will tell you the values that are unique for ex: yes and no
- pd.nunique() will give a count of how many unqiue values
- pd.value_counts() each count of number of unique values ex: no- 50, yes -70
- pd.sort_index() will sort the index alphabetically or numerically
- pd.sort_values() will sort based on the values
- pd.apply() will apply any function you create [Broadcasting a fucntion to a column]
- pd.to_datetime('column') use to convert categorical data in date and time data
