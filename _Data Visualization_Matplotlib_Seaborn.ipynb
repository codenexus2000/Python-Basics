```python
#Making a graph for most selling items
graph_data3= data.groupby('Product Name').agg({'Sales':sum}).sort_values(by='Sales', ascending=False).head(10)

new_index=[]

for value in graph_data3.index:
    new_index.append(value[:20])
   
graph_data3.index= new_index

fig,my_ax= plt.subplots(figsize=(10,4))

graph_data3.plot.bar(ax=my_ax)
plt.xticks(rotation=60)
plt.title('Most Order Items')
plt.ylabel('Sales ($)')
plt.grid(linestyle= ':')
```

```python
graph_data4= (data.groupby('Sub-Category')
              .agg({'Sales':sum, 'Profit':sum})
              .sort_values(by='Sales', ascending= False))

fig,my_ax= plt.subplots(figsize=(10,5))
graph_data4.plot.bar(ax=my_ax)
plt.title("Sales-Profit Per Sub Category")
plt.ylabel("Sales/Profit")
plt.xlabel(None)
plt.grid(linestyle=':')
plt.tight_layout()
```

```python
#making a stacked bar graph for catergory sales per regions

graph_data5= (data
              .groupby(['Region','Category'])
              .agg({'Sales':sum})
              .unstack(0)) #unstack makes a table based on column number (0,1,2)

fig,my_ax= plt.subplots(figsize=(10,7))
graph_data5.plot(ax=my_ax, kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.ylabel('Sales ($)')
plt.title('Sales per Category in each Region')
plt.tight_layout()
```

```python
#making a stacked bar graph for catergory sales per regions

graph_data6= (data
              .groupby(['Region','Category'])
              .agg({'Sales':sum})
              .unstack(1)) #unstack makes a table based on column number (0,1,2)

fig,my_ax= plt.subplots(figsize=(10,7))
graph_data6.plot(ax=my_ax, kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.ylabel('Sales ($)')
plt.title('Sales per Category in each Region')
plt.tight_layout()
```

# Pareto Analysis

```python
cx_profit= data.groupby('Customer ID').agg({'Profit':sum}).sort_values('Profit', ascending= False)

total_profit= cx_profit['Profit'].sum()

percentage_profit= cx_profit/total_profit
percentage_profit=percentage_profit.cumsum()

percentage_profit['Count']= range(1,805)
percentage_profit= percentage_profit.set_index('Count')
```

```python
cx_sales= data.groupby('Customer ID').agg({'Sales': sum}).sort_values('Sales', ascending= False)
cx_sales

total_sales= cx_sales['Sales'].sum()
total_sales

percentage_sales= cx_sales/total_sales
percentage_sales= percentage_sales.cumsum() #cumsum gives a cummlative combinations of the data

percentage_sales['Count']= range(1,805)
percentage_sales= percentage_sales.set_index('Count')
```

```python
fig, my_ax= plt.subplots(figsize=(14,5))
percentage_sales.plot(ax=my_ax)
percentage_profit.plot(ax=my_ax)
plt.title('Pareto Analysis') #pareto analysis can help understand the most performing and least performing assests for an buissness

plt.ylabel('Percentage for Sales')
plt.xlabel('Number of Custome')
plt.grid(linestyle= ':')
plt.tight_layout()
```
