```python
yaml_content = """path: C:/Users/adi20/Desktop/dataset
train: images
val: images

nc: 2
names: ["class1", "class2"]  # Replace with actual class names
"""

file_path = "C:/Users/adi20/Desktop/dataset/dataset.yaml"

# Create the dataset.yaml file
with open(file_path, "w") as file:
    file.write(yaml_content)

print(f"dataset.yaml created at: {file_path}")
```

```python
from ultralytics import YOLO


model = YOLO("yolov8n.pt") 

model.train(data="C:/Users/adi20/Desktop/dataset/dataset.yaml",
            epochs=50, 
            imgsz=640,
            batch=1) 
```

```python
import os

base_path = "runs/detect/"
for root, dirs, files in os.walk(base_path):
    if "best.pt" in files:
        print(f"✅ Model found at: {os.path.join(root, 'best.pt')}")
        break
else:
    print("❌ No trained model (best.pt) found!")
```

```python
model = YOLO("runs/detect/train7/weights/best.pt")


results = model("kirana_store.jpg", save=True)

print(results)
```

```python
import cv2
import matplotlib.pyplot as plt

image_path = "kirana_store.jpg" 
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

model = YOLO("runs/detect/train7/weights/best.pt")

results = model(image)

# bounding box
for result in results:
    for box in result.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box

#displaying:
plt.figure(figsize=(10,6))
plt.imshow(image)
plt.axis("off")
plt.title("Detected Objects in Kirana Store")
plt.show()
```
